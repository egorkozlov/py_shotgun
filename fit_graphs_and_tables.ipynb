```python
# part 7.1 doubling social stigma

# table

from tiktak import filer
from fit_plot import FitPlots

import numpy as np
from matplotlib import pyplot as plt

diff = False


sm_25 = r'\textit{...at 25} '
sm_35 = r'\textit{...at 35} '

em_25 = r'\textit{...at 25} '
em_35 = r'\textit{...at 35} '

im_30 = r'\textit{...at 30} '
im_40 = r'\textit{...at 40} '


ed_30= r'\textit{...at 30} '
ed_40 = r'\textit{...at 40} '

dm_10 = r'\textit{...marriage-first} '
dk_10 = r'\textit{...kids-first} '


wk_25 = r'\textit{...at 25} '
wk_35 = r'\textit{...at 35} '

wc_m = r'\textit{median male at 23} '
wc_f = r'\textit{median female at 21} '

ck_md = r'\textit{child ever born, median} '
ck_30 = r'\textit{child born at 30, median} '

a = r'\textit{\% aborted} '


def welfare_return(bl_name,cf_name):
    fp = FitPlots(targ_mode='college',
                   compare=bl_name,
                   base=cf_name,noplot=True)
    
    a_fem, a_mal, _ = fp.plot_welfare(noplot=True)
    return a_fem, a_mal




for educ, name in zip(['hs','col'],['high school graduates','college graduates']):

    name_bl = '{} baseline.pkl'.format(educ)
    name_ca = '{} costless abortion.pkl'.format(educ)
    name_ns = '{} no social stigma.pkl'.format(educ)
    name_cans = '{} abortions and no stigma.pkl'.format(educ)
    
    for name in (name_bl, name_ca, name_ns, name_cans):
        
        mom = filer(name,0,0)
        
        w_fem, w_mal = welfare_return(name_bl,name)
        
        
        sm_25 += r' & ' + '{:02.1f}'.format(100*mom['single mothers in total at 25'])
        sm_35 += r' & ' + '{:02.1f}'.format(100*mom['single mothers in total at 35'])
        
        em_25 += r' & ' + '{:02.1f}'.format(100*mom['ever married at 25'])
        em_35 += r' & ' + '{:02.1f}'.format(100*mom['ever married at 35'])
        
        im_30 += r' & ' + '{:02.1f}'.format(100*(1-mom['never married and kids in population at 30'] - mom['never married and no kids in population at 30'] - mom['divorced and no kids in population at 30']  - mom['divorced and kids in population at 30']))
        im_40 += r' & ' + '{:02.1f}'.format(100*(1-mom['never married and kids in population at 40'] - mom['never married and no kids in population at 40'] - mom['divorced and no kids in population at 40']  - mom['divorced and kids in population at 40']))
        
        ed_30 += r' & ' + '{:02.1f}'.format(100*mom['ever divorced at 30'])
        ed_40 += r' & ' + '{:02.1f}'.format(100*mom['ever divorced at 40'])
        
        dm_10 += r' & ' + '{:02.1f}'.format(100*mom['divorced by years after marriage if marriage first, 10'])
        dk_10 += r' & ' + '{:02.1f}'.format(100*mom['divorced by years after marriage if kids first, 10'])
        
        wk_25 += r' & ' + '{:02.1f}'.format(100*mom['with kids at 25'])
        wk_35 += r' & ' + '{:02.1f}'.format(100*mom['with kids at 35'])
        
        wc_f += r' & ' + '${:02.1f}$'.format(w_fem[3])
        wc_m += r' & ' + '${:02.1f}$'.format(w_mal[2])
        
        
        ck_md += r' & ' + '{:02.1f}'.format(mom['welfare of all children born, median'])
        ck_30 += r' & ' + '{:02.1f}'.format(mom['child welfare if born at 30, median'])
        
        a += r' & ' + '{:02.1f}'.format(100*mom['unplanned pregnancies aborted'])
        
        
        
print('single mothers in population:')
#print(sm_25)
print(sm_35)
print('ever married:')
print(em_25)
print(em_35)
print('is married:')
print(im_30)
print(im_40)
print('ever divorced:')
#print(ed_30)
print(ed_40)
print('with kids:')
print(wk_25)
print(wk_35)
print('divorced 10 years after if...')
print(dm_10)
print(dk_10)
print('welfare comparisons:')
print(wc_f)
print(wc_m)
print('childs consumption:')
print(ck_md)
print(ck_30)
print(a)
```
