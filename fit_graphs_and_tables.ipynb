```python
print('\n\n\n Table with fit \n\n\n')
from tiktak import filer
mom_col = filer('col baseline.pkl',0,0)
mom_hs = filer('hs baseline.pkl',0,0)
from targets import all_targets
tar_col = all_targets(t_col)
tar_hs = all_targets(t_hs)



def getkey(key):
    return [100*mom_col[key],100*tar_col[key][0],100*mom_hs[key],100*tar_hs[key][0]]
def getcell(key):
    return '{:02.1f} & {:02.1f} & {:02.1f} & {:02.1f}'.format(*getkey(key))

def getrow(desc,key):
    return r'{\footnotesize ' + desc + '} & '+ getcell(key) + r' \\'

print(getrow(r' \% with more than one marriage at 40','more than one mar at 40'))
print(getrow(r'\% in labor force at 30 if married and kids','in labor force at 30 if kids'))
print(getrow(r'mean \% of child expenditures in total earnings','mean x share'))
print(getrow(r'\% unplanned pregnancies aborted','unplanned pregnancies aborted'))
print(getrow(r'abortions in 30s as \% of abortions in 20s','abortion 30s over 20s'))
```
